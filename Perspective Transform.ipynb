```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def transform(image, original_pts, desired_pts, title_list):
    cv2.circle(image, tuple(original_pts[0]), 5, (0, 0, 255), -1)
    cv2.circle(image, tuple(original_pts[1]), 5, (0, 0, 255), -1)
    cv2.circle(image, tuple(original_pts[2]), 5, (0, 0, 255), -1)
    cv2.circle(image, tuple(original_pts[3]), 5, (0, 0, 255), -1)
    
    matrix = cv2.getPerspectiveTransform(original_pts, desired_pts)
    result = cv2.warpPerspective(image, matrix, (1024, 768))

    fig = plt.figure(figsize=(13, 9))

    fig.add_subplot(1,2,1)
    plt.title(title_list[0])
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    fig.add_subplot(1,2,2)
    plt.title(title_list[1])
    plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.tight_layout()
    plt.show()
    
image_list = []
title_list = []
for i in range(1,5):
    image_list.append('./data/dsu' + str(i) + '.jpg')
    title_list.append(['DSU ' + str(i) + ' : Original Image','DSU ' + str(i) + ' : Perspective Transform Result'])
```

```python
img = cv2.imread(image_list[0])
original_pts = np.float32([[322, 279], [759, 311], [325, 535], [766, 508]])
desired_pts = np.float32([[322, 279], [759, 279], [322, 535], [759, 535]])
transform(img, original_pts, desired_pts, title_list[0])
```
